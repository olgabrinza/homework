## Классификация картинок

```python
import pandas as pd
from sklearn.datasets import load_digits
```

```python
data = load_digits()
```

```python
data.keys()
```

```python
# описание данных
for line in data['DESCR'].split('\n'):
    print( line )
```

```python
# закодированные признаки

X = pd.DataFrame( data['data'] )
X.head()
```

```python
# целевая переменная - цифра от 0 до 9

Y = pd.Series( data['target'] )
Y.head()
```

```python
Y.tail()
```

```python
Ydummy = pd.get_dummies( Y )
Ydummy.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split( X, Ydummy, test_size = 0.3, random_state = 0 )
```

### Обучаем дерево, выбираем лучшие параметры

```python
from sklearn.tree import DecisionTreeClassifier
import numpy as np
```

```python
from sklearn.grid_search import GridSearchCV
```

```python
parameters = {'min_samples_split':np.arange(2, 10), 'max_depth': np.arange(1,10), 'criterion':['gini', 'entropy'],
              'min_samples_leaf' :np.arange(1, 10)}
```

```python
clfr = DecisionTreeClassifier()
grid = GridSearchCV(clfr, parameters,scoring='accuracy', cv=5, verbose =5)
```

```python
grid.fit(X,Ydummy)
```

```python
print('The parameters combination that would give best accuracy is : ')
print(grid.best_params_)
print('The best accuracy achieved after parameter tuning via grid search is : ', grid.best_score_)
```
